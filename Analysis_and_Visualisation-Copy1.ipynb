```python
def sg_ott_apr_mean(a):      # Defining custom function sg_ott_apr_mean that can be used to get the mean of the two columns for each golfer and add it as another column
    ott_app_m = a[['SG:OTT','SG:APR']].mean()
    return round(ott_app_m,2)

pga.apply(sg_ott_apr_mean, axis = 1).dropna() # Calling the sg_ott_apr_mean function on the pga dataset
```

```python
avg_driving_distance = pga[['Player', 'AVG_Driving_DISTANCE', 'SG:APR', 'SG:OTT']].dropna() # Creating avg_driving_distance DataFrame

avg_driving_distance['MEAN_SG_APR_OTT'] = avg_driving_distance.apply(sg_ott_apr_mean, axis=1) # Calling the sg_ott_apr_mean function to add the MEAN_SG_APR_OTT column to avg_driving_distance 

top_50_driving_distance = avg_driving_distance.sort_values('AVG_Driving_DISTANCE', ascending = False).head(50) # Create top_50_driving_distance DataFrame of top 50 golfers in AVG_Driving_DISTANCE with longest AVG_Driving_DISTANCE on top
top_50_driving_distance
```

```python
bottom_50_driving_distance = avg_driving_distance.sort_values('AVG_Driving_DISTANCE', ascending = False).tail(50) # Create top_50_driving_distance DataFrame of bottom 50 golfers in AVG_Driving_DISTANCE by ordering in descending manner and selecting bottom 50
bottom_50_driving_distance # Show bottom_50_driving_distance DataFrame
```

```python
round(top_50_driving_distance['MEAN_SG_APR_OTT'].mean(), 2)
# Aim is to see if the top 50 golfers in driving distance had better a better score for the mean of SG:OTT (Strokes Gained Off The Tee) and SG:APR (Strokes Gained Approach) when compared to the bottom 50 golfers in driving distance 
```

```python
round(bottom_50_driving_distance['MEAN_SG_APR_OTT'].mean(), 2)
# When looking at the mean of SG:OTT (Strokes gained off the tee) and SG:APR (Strokes Gained Approach), it's clear that driving distance helps as the top 50 golfers in average driving distance have a much better score for MEAN_SG_APR_OTT (mean of SG:OTT and SG:APR) than the bottom 50 golfers in average driving distance.
```

```python
pga_top_10s = pga[['Player', 'NUMBER_OF_TOP_Tens']] # Create DataFrame

pga_top_10s['Over_2_Top_10s'] = pga_top_10s['NUMBER_OF_TOP_Tens'].apply(lambda x: 'Yes' if x > 2.0 else 'No') # Conditional statement adding Over_2_Top_10s column. Value will be Yes if golfer has more than 2 top 10s or No if golfer has 2 or less top 10s. 
print(pga_top_10s[['Player', 'NUMBER_OF_TOP_Tens', 'Over_2_Top_10s']]) # Print result
```
